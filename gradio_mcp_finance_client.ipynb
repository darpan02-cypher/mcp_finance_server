```python
from groq import Groq
from google.colab import userdata
```

```python
client = Groq(api_key=userdata.get("GROQ_API_KEY"))
MODEL = "openai/gpt-oss-120b"
```

```python
# MCP Server configuration for Finance server
MCP_FINANCE_SERVER = {
    "type": "mcp",
    "server_label": "Finance",
    "server_url": "https://56f2e893bb4237a5ff.gradio.live/gradio_api/mcp/"
}
```

```python
def query_finance_with_mcp(user_query: str):
    """
    Query the Finance MCP server using Groq

    Args:
        user_query: The user's question about stocks/finance
        model: Groq model to use (default: mixtral-8x7b-32768)

    Returns:
        The assistant's response with financial data
    """
    try:
        completion = client.chat.completions.create(
            model=MODEL,
            messages=[
                {
                    "role": "user",
                    "content": user_query
                }
            ],
            tools=[MCP_FINANCE_SERVER],
        )

        return completion.choices[0].message
    except Exception as e:
        return f"Error: {str(e)}"
```

```python
query1 = "What is the current price of Apple stock (AAPL)?"
print(f"\nQuery: {query1}")
response1 = query_finance_with_mcp(query1)
print(f"\nResponse:")
print(response1.content)
```

```python
query2 = "Compare the stocks AAPL, MSFT, and TSLA. Show me their prices and PE ratios."
print(f"\nQuery: {query2}")
response2 = query_finance_with_mcp(query2)
print(f"\nResponse:")
print(response2.content)
```

```python
query3 = "What has been the price trend for Tesla (TSLA) over the last 30 days? Show me the start price, end price, high, low, and percent change."
print(f"\nQuery: {query3}")
response3 = query_finance_with_mcp(query3)
print(f"\nResponse:")
print(response3.content)
```
